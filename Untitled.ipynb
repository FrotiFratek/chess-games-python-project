```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

chess_data = pd.read_csv('games.csv')
#first look at dataset and data types
print(chess_data.head())
print(chess_data.info())

#checking winrate of white (taking in consideration only games where someone has won)
print(len(chess_data[chess_data.winner == 'white']) / len(chess_data[chess_data.winner != 'draw']))
#0.5233933431023655
```

```python
#checking if there is an linear association between mean of rank of opponents and turns
chess_data['mean_rank'] = chess_data.apply(lambda row: (row['white_rating'] + row['black_rating']) / 2, axis=1)
plt.figure(figsize=(20,20))
plt.scatter(x = chess_data.mean_rank, y = chess_data.turns)
plt.xlabel('Mean of Opponents Rank')
plt.ylabel('Number of Turns')
plt.show()

from scipy.stats import pearsonr
corr_rank_turns, p = pearsonr(chess_data.mean_rank, chess_data.turns)
print(corr_rank_turns)
```

```python
#countplot of victory status
print(chess_data.victory_status.value_counts(normalize=True))

import seaborn as sns
sns.countplot(x='victory_status', data=chess_data)
plt.show()
plt.close()
```
